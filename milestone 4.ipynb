Q.How do dietary preferences (diet) distribute across the platform? 

Purpose:To check the distribution of different types of people based on their diet preferences.

```python
# 1.Define a function group_diet
def group_diet(diet):
    if pd.isna(diet): return 'Not Specified'
    diet = diet.lower()
    if 'anything' in diet: return 'Anything'
    elif 'vegetarian' in diet: return 'Vegetarian'
    elif 'vegan' in diet: return 'Vegan'
    elif 'kosher' in diet: return 'Kosher'
    elif 'halal' in diet: return 'Halal'
    else: return 'Other'

df['diet_group'] = df['diet'].apply(group_diet)

# 2. Calculate percentages for users who provided an answer
df_responded = df[df['diet_group'] != 'Not Specified']
diet_proportions = df_responded['diet_group'].value_counts(normalize=True) * 100
```

```python
diet_proportions
```

Q.Plot a histogram of age with a vertical line indicating the mean age. What does the distribution reveal about the most common age group on the platform

```python
# 1. Calculate key metrics
mean_age = df['age'].mean()
median_age = df['age'].median()

# 2. Create the histogram
plt.figure(figsize=(12, 6))
sns.histplot(df['age'], bins=30, kde=True, color='teal', edgecolor='black')

# 3. Add vertical lines for mean and median
plt.axvline(mean_age, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_age:.2f}')
plt.axvline(median_age, color='orange', linestyle='-', linewidth=2, label=f'Median: {median_age:.2f}')

# 4. Formatting
plt.title('Distribution of User Age', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Number of Users', fontsize=12)
plt.legend()
plt.grid(axis='y', alpha=0.3)
```

Analysis:The distribution is heavily "right-skewed," meaning the majority of users are clustered at the younger end of the scale. 
         The most common age group consists of individuals in their mid-to-late 20s and early 30s.
         There is a massive peak between ages 22 and 35. 
         This suggests that the platform’s target demographic is predominantly young professionals and post-graduates.

Q.How does the age distribution differ by gender?

```python
# 1. Calculate statistics
stats = df.groupby('gender')['age'].describe()
print(stats)

# 2. Create the visualization
plt.figure(figsize=(12, 6))

# Plotting the overlapping density curves
sns.kdeplot(data=df, x='age', hue='gender', fill=True, common_norm=False, 
            palette=['lightcoral', 'skyblue'], alpha=0.5, linewidth=2)

# Adding vertical lines for the mean of each gender
plt.axvline(df[df['gender'] == 'f']['age'].mean(), color='darkred', linestyle='--', label='Mean (f)')
plt.axvline(df[df['gender'] == 'm']['age'].mean(), color='darkblue', linestyle='--', label='Mean (m)')

# 3. Formatting
plt.title('Age Distribution Density by Gender', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend()
plt.grid(axis='y', alpha=0.3)
```

Analysis:Both genders have a median age of 30, meaning that for both men and women, exactly half of the user base is 30 or younger.
         Both distributions are "right-skewed," with the highest density of users found between ages 24 and 34.
         The slightly higher mean and standard deviation for women suggest 
         that there is a marginally higher proportion of women in the 40+ age groups compared to men.

Analysis:The visualization reveals that there is no significant linear relationship between age and income on the platform. 
         Unlike the general economic expectation where income increases with professional 
         experience (and thus age), the users in this dataset report a wide range of incomes across all age groups.

Q.Create boxplots of income grouped by age_group

```python
#Q.Analyze income levels within gender and status categories.

# 1. Group by gender and status for mean/median calculations
stats = df.groupby(['gender', 'status'], observed=True)['income'].agg(['mean', 'median']).reset_index()

# 2. Visualize Average Income
plt.figure(figsize=(12, 6))
sns.barplot(x='status', y='mean', hue='gender', data=stats, palette=['lightcoral', 'skyblue'])

plt.title('Average Income by Relationship Status and Gender', fontsize=14)
plt.ylabel('Average Income ($)')
plt.xlabel('Relationship Status')
plt.legend(title='Gender')
plt.grid(axis='y', linestyle='--', alpha=0.5)

# 3. Create pivot table for summary
pivot_summary = df.pivot_table(values='income', index='status', columns='gender', aggfunc='median')
print("Median Income Summary:")
print(pivot_summary)
```

```python
#Q.How do pets preferences vary across gender and age_group?
# 1.Define bins and age_group
bins = [18, 25, 35, 45, 55, 70]
labels = ['18-24', '25-34', '35-44', '45-54', '55-69']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)

# 2. Define pet sentiment
def get_sentiment(val):
    if pd.isna(val): return 'None'
    return 'Dislike' if 'dislikes' in val.lower() else 'Like/Have'

df['pet_sentiment'] = df['pets'].apply(get_sentiment)

# 3. Calculate Disclosure Rate (Participation)
disclosure = df.groupby(['age_group', 'gender'], observed=True)['pets'].apply(lambda x: x.notna().mean() * 100).unstack()

# 4. Calculate Sentiment (Among those who disclosed)
df_resp = df[df['pet_sentiment'] != 'None']
sentiment = df_resp.groupby(['age_group', 'gender'], observed=True)['pet_sentiment'].apply(lambda x: (x == 'Like/Have').mean() * 100).unstack()

# 5. Visualization (disclosure rate)
disclosure.plot(kind='bar', figsize=(10, 6), color=['lightcoral', 'skyblue'], edgecolor='black')
plt.title('Pet Preference Disclosure Rate by Age/Gender')
plt.ylabel('Percentage Disclosure (%)')
```

Q.How does sign vary across gender and status?

```python

# 1. Analyze Variation by Gender
# Create a normalized cross-tabulation to see the % of each sign within each gender
sign_gender_ct = pd.crosstab(df['sign_clean'], df['gender'], normalize='columns') * 100

# 2. Analyze Variation by Status
# Create a normalized cross-tabulation to see the % of each sign within each status group
sign_status_ct = pd.crosstab(df['sign_clean'], df['status'], normalize='columns') * 100

# 3. Visualization: Heatmap for Gender
plt.figure(figsize=(10, 8))
sns.heatmap(sign_gender_ct, annot=True, cmap='YlGnBu', fmt=".1f")
plt.title('Zodiac Sign Proportions by Gender (%)', fontsize=14)
plt.ylabel('Zodiac Sign')
plt.xlabel('Gender')
plt.tight_layout()
plt.savefig('sign_gender_analysis.png')

# 4. Visualization: Heatmap for Status
plt.figure(figsize=(12, 8))
sns.heatmap(sign_status_ct, annot=True, cmap='YlGnBu', fmt=".1f")
plt.title('Zodiac Sign Proportions by Relationship Status (%)', fontsize=14)
plt.ylabel('Zodiac Sign')
plt.xlabel('Status')

# Display result summary
print("\nSign Distribution by Gender (%):")
print(sign_gender_ct.round(2))
```
